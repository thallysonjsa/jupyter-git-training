```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('../data/winequality-red.csv',sep = ';')
df2 = pd.read_csv('../data/winequality-white.csv', sep=';')

df = pd.concat([df, df2])
df.head()
```

```python
x = df['volatile acidity']
y = df['citric acid']

plt.figure(figsize=(12,6))
plt.scatter(x, y)
plt.show()
```

```python
x = df['quality'].unique()
y = df.groupby('quality').agg({'citric acid': 'mean'})['citric acid']

plt.figure(figsize=(12,6))
plt.bar(x, y)
plt.show()
```

```python
x = df['quality']

plt.figure(figsize=(12,6))
plt.hist(x, bins=7)
plt.show()
```

```python
x = ['20/09', '21/09', '22/09', '23/09', '24/09']
y = df.iloc[:5]['citric acid']
y2 = df.iloc[:5]['quality']

plt.figure(figsize=(12,6))
plt.plot(x, y)
plt.plot(x, y2)
plt.xticks(x)
plt.show()
```

```python
x = df['citric acid']

plt.figure(figsize=(8,6))
plt.boxplot(x)
plt.savefig('../reports/boxplot-exemplo.png')
```
